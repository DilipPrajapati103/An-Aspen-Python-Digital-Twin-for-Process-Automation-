```python
combined_df  = pd.read_excel(r"C:\Users\ravik\Desktop\CSTR digital twin\data\Raw_data.xlsx")
combined_df
```

```python
combined_df = combined_df.drop(["Unnamed: 0"],axis= 1)
```

```python
null_row = []
for i in range(combined_df["PDO_fraction"].shape[0]):
    
    if combined_df["PDO_fraction"].isnull()[i] == True:
         null_row.append(i)  
    
    else :
        continue
   
```

```python
null_rows = np.array(null_row)
```

```python
combined_df = combined_df.drop(null_rows)  
combined_df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x= combined_df["R_temp"]
y= combined_df["PDO_fraction"]
plt.scatter(x,y)
plt.title("R_Temp vs PDO_fraction")
plt.show()
```

```python
x= combined_df["R_vol"]
y= combined_df["PDO_fraction"]
plt.scatter(x,y)
plt.title("R_vol vs PDO_fraction")
plt.show()
```

```python
# 1d series to 2D array
z= np.array(combined_df["PDO_fraction"]).reshape(-1,1)
```

```python
x= combined_df["R_vol"]
y= combined_df["R_temp"]
z= np.array(combined_df["PDO_fraction"]).reshape(-1,1)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111,projection='3d')

ax.plot_surface(x,y,z)

plt.show()
```

## Correlation Plot

```python
corr = combined_df.corr(numeric_only=True)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```
